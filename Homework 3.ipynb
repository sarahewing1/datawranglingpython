I'm going to quickly restrict the data and only look at Gen1 Pokemon without the "Mega" label for simplicity.

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sb
import numpy as np

df = pd.read_csv("Pokemon.csv")
# Filtering on the data. Don't get caught up on this, we'll cover it very soon!
df = df[df.Generation == 1 & ~df.Name.str.contains("Mega")]
#df2 = df[df.Name.str.contains("Mega")]
df.head()
#df2.head()
```

## Pokemon Attack and Defense Correlation

```python
df.plot.scatter("Attack", "Defense",s=5);
sb.lmplot("Attack", "Defense", hue="Type 1", data=df, fit_reg=False);#seaborn is better here because I can color by the Type
```

## Strongest Pokemon by Summed Stats

```python
df.plot.bar("Name", "Total");#hard to read
df.sort_values("Total", ascending=False).iloc[:20, :].plot.bar("Name", "Total");
#df2.sort_values("Total", ascending=False).iloc[:20, :].plot.bar("Name", "Total");
```
